```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
#import matplotlib.pyplot as plt

#Sample dataset
dataset = ('C:/Users/DELL/Downloads/internship/bank.csv')
#loading the dataset into df
df = pd.read_csv(dataset, sep=';')
(df)
```

```python
##Preprocess the Data
# Check for missing values
print(df.isnull().sum())

# Convert categorical variables to numerical values
df_encoded = pd.get_dummies(df, drop_first=True)

# Separate features and target variable
X = df_encoded.drop('y_yes', axis=1)
y = df_encoded['y_yes']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
try:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    print(f"Training data shape: {X_train.shape}")
    print(f"Testing data shape: {X_test.shape}")
    print(f"Training labels shape: {y_train.shape}")
    print(f"Testing labels shape: {y_test.shape}")
except Exception as e:
    print(f"Error occurred: {e}")
```

```python
# Import necessary libraries
from sklearn.tree import DecisionTreeClassifier

# Initialize the model
clf = DecisionTreeClassifier(random_state=42)

# Train the model
clf.fit(X_train, y_train)

# Output the trained model (optional)
print(clf)
```

```python
print(f"Number of features: {clf.n_features_in_}")
print(f"Depth of the tree: {clf.get_depth()}")
print(f"Number of nodes: {clf.get_n_leaves()}")
```

```python
##Evaluate the Model
# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

print('Classification Report:')
print(classification_report(y_test, y_pred))

print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```
